## Get âˆ†T for each month of each year

- Group by time.year and apply subtraction for each month

```python
clim_t = era_climatology.temperature_2m.copy()

def sub_months(era_t_yr):
    clim_t['time'] = era_t_yr.time
    return era_t_yr - clim_t
    
delta_t = era_weather.temperature_2m.groupby('time.year').apply(sub_months)
delta_t = xr.Dataset({'delta_t_2m': delta_t},
                     coords={'longitude': delta_t.longitude,
                    'latitude': delta_t.latitude,
                    'time': delta_t.time
                    }
          )
```

## Get the mean delta for each year

```python
year_mean = delta_t.resample('AS', dim='time')
```

```python
year_mean
```

### Reproject the yearly delta values onto the population grid

```python
from rasterio.warp import reproject, Resampling
import population_tools
```

```python
height, width = population_tools.get_shape()
pop_crs, era_affine = population_tools.get_era_compat_crs_affine()
pop_shape = population_tools.get_shape()
```

```python
year = 2000
```

```python
def project_to_population(year, weather_param):
    print(year)
    newarr = np.empty(shape=pop_shape)
    reproject(
        weather_param.values, newarr,
        src_transform=common_affine,
        dst_transform=era_affine,
        src_crs=pop_crs,
        dst_crs=pop_crs,
        resample=Resampling.cubic_spline)

    # Wrap as DataArray to make sure the coordinates line up
    resamp_dt = xr.DataArray(newarr, coords={'latitude': pop_data.latitude, 
                                             'longitude': pop_data.longitude}, 
                 dims=['latitude', 'longitude'])

    pop_year = pop_data.population.sel(time=f'{year}')
    
    res = pop_year * resamp_dt
    return res
```

```python
exposure = project_to_population(year, year_mean.delta_t_2m.sel(time=f'{year}'))
```

```python
years = list(range(2000, 2016))
```

```python
yr_exposures = []
for year in years:
    print(year)
    exposure = project_to_population(year, year_mean.delta_t_2m.sel(time=f'{year}'))
    yr_exposures.append(exposure)

```

```python
yr_exposures = xr.concat(yr_exposures, dim='time')
```

```python
yr_exposures.name = 'year_mean_exposure'
yr_exposures.to_netcdf(str(DATA_SRC /'lancet' / 'year_mean_exposure.nc'))
```
